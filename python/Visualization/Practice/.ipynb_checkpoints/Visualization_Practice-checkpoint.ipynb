```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x, y)
```

```python
x1 = [1,2,3,4,5]
y1 = [i ** 2 for i in x]

x2 = [1,2,3,4,5]
y2 = [6,7,8,9,10]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
x1 = range(50)
y1 = [np.cos(value) for value in x1]

x2 = range(50)
y2 = [np.sin(value) for value in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
plt.figure(figsize = (10, 5))

plt.subplot(1, 3, 1)
plt.plot(x1, y1, c = 'r', linestyle = 'dotted')

plt.subplot(1, 3, 2)
plt.plot(x2, y2, c = 'g', linestyle = 'dashed')

plt.subplot(1, 3, 3)
plt.plot(x1, y2, c = 'b', linestyle = 'dashed')

```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value + 50 for value in x2]

plt.plot(x1, y1, c = 'r', ls = '-.', label = 'Line1')
plt.plot(x2, y2, c = 'y', ls = ':', label = 'Line2')

plt.text(40, 30, 'Line1')
plt.text(60, 100, 'Line2')
```

```python
n = 50

x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)
area = np.pi* (30 * np.random.rand(n)) ** 2

plt.scatter(x,y, c = colors, s = area, alpha = 0.2)
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x, y, color = 'r', width = 0.5)
```

```python
data = np.array([[5, 25, 50, 20], [4, 23, 51, 17], [6, 22, 52, 19]])

plt.bar(x, data[0], color = 'b', width = 0.5)
plt.bar(x, data[1], color = 'g', width = 0.5, bottom = np.sum(data[:1], axis = 0))
```
