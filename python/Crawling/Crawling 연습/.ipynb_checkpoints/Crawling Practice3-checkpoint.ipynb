```python
html = """
<body>
<div>
    <div class='snack'>
        <p id='first'> 양파링</p>
        <p id='second'> 새우깡</p>
        <p> 초코파이</p>
        <p class='sweet'> 맛동산 </p>
        
    <div class='icecream'>
        <p> 빵빠레</p>
        <p id='third'> 죠스바</p>
        <p> 꼬깔콘</p>
        <p class='sweet'> 쭈쭈바</p>
    </div>
</div>
</body>
"""
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
# p태그의 텍스트를 모두 출력하시오..

pls = soup.select('p')

for pl in pls:
    print(pl.text)
```

```python
# '양파링'을 출력하시오..
pls[0].text
```

```python
# '죠스바'를 출력하시오..
soup.select('p#third')[0].text
```

```python
# '맛동산'을 출력하시오..
soup.select('p.sweet')[0].text
```

```python
res = requests.get('https://www.naver.com/')
```

```python
# 다음 뉴스 타이틀 크롤링..

url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EC%97%98%EB%93%A0%EB%A7%81'

res = requests.get(url)
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
path = '#newsColl > div.cont_divider > ul > li'
els = soup.select(path)
els
```

```python
for el in els:
    print(el.select('a.tit_main')[0].text)
    print(el.select('a.tit_main')[0]['href'])
    print(el.select('p.desc')[0].text)
    print('-----------------------------------')
    print()
```
