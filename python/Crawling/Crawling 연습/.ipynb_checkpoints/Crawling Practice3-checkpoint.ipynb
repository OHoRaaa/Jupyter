```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
url = 'https://news.naver.com/main/read.naver?mode=LPOD&mid=sec&oid=001&aid=0012930410&isYeonhapFlash=Y&rc=N'

driver.get(url)
```

```python
html = driver.page_source
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
path = '#cbox_module_wai_u_cbox_content_wrap_tabpanel > ul > li'

cos = soup.select(path)
cos
```

```python
for co in cos:
    print('Time', co.select('span.u_cbox_date')[0].text)
    print('Contents', co.select('span.u_cbox_contents')[0].text)
    print('Like', co.select('em.u_cbox_cnt_recomm')[0].text, '/', 'Hate', co.select('em.u_cbox_cnt_unrecomm')[0].text)
    print()
    
```
