```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.amazon.com/bestsellers')
```

```python
driver.find_element_by_link_text('Appliances').click()
```

```python
for i in range(1,3):
    if i == 2:
        url = 'https://www.amazon.com/Best-Sellers-Appliances/zgbs/appliances/ref=zg_bs_pg_{}?_encoding=UTF8&pg=2'
        url = url.format(2)
        
        driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    lis = soup.select('#zg-ordered-list > li')
#     print(lis)
    for li in lis:
        try:
            print(li.select('#zg-ordered-list > li > span > div > div > span.a-size-small.aok-float-left.zg-badge-body.zg-badge-color > span')[0].text)
            print(li.select('#zg-ordered-list > li > span > div > span > a:nth-child(1) > div')[0].text)
            print(li.select('#zg-ordered-list > li > span > div > span > div > a.a-size-small.a-link-normal')[0].text)
            print(li.select('span.a-size-base')[0].text)
            print()
        except:
            pass
    
```
