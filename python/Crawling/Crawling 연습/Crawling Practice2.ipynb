```python
# 다음 뉴스 타이틀 크롤링..

url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EC%97%98%EB%93%A0%EB%A7%81'
res = requests.get(url)
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
path = '#newsColl > div.cont_divider > ul > li'
soup.select(path)
```

```python
tags = soup.select(path)
for tag in tags:
    print(tag.select('a.tit_main')[0].text)
    print(tag.select('a.tit_main')[0]['href'])
    print(tag.select('p.desc')[0].text)
    print('----------')
```

```python
# 검색어를 입력 받은 후 해당 검색어로 검색한 결과 다음 뉴스 페이지에서 
# 검색한 결과의 제목, url, summary를 크롤링하시오
# (url을 잘 조사하면... )
```

```python
url = 'https://search.daum.net/search?nil_suggest=btn&w=news&DA=SBC&q={}'
search = input('검색어를 입력하세요')

```

```python
url = url.format(search)
res = requests.get(url)
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')

path = '#newsColl > div.cont_divider > ul > li'
soup.select(path)
```

```python
tags = soup.select(path)

for tag in tags:
    print(tag.select('a.tit_main')[0].text)
    print(tag.select('a.tit_main')[0]['href'])
    print(tag.select('p.desc')[0].text)
    print('---------------------------')
    print()
```
