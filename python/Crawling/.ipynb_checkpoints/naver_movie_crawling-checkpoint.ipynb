```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://movie.naver.com/movie/point/af/list.naver?&page=1'

res = requests.get(url)
res.text

```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
path1 = 'td.title > a.movie.color_b'
titles = soup.select(path1)
titles
```

```python
path2 = ' td.title > div > span > span'
star = soup.select(path2)
star
```

```python
path3 = 'td.title'
soup.select(path3)[0]
```

```python
for t in titles:
    t.text
    
for s in star:
    print(s.select('span'))
```
