```python

df['분양가격(제곱미터)'].apply(lambda x : str(x).strip().replace(',', ''))
```

```python
indexes = []
errors = []
for i,j in enumerate(df['분양가격(제곱미터)']):
    try:
        float(j)
        indexes.append(i)
    except:
        errors.append(j)
```

```python
errors
```

```python
df_clean = df.iloc[indexes]
df_clean
```

```python
# 2021년도의 지역별 분양가 평균
df_clean.groupby('지역명')['분양가격(제곱미터)'].mean()
```

```python
# 2015 ~ 2021년 사이 지역별 분양가 변화를 line chart로 그리시오
```

```python
df_clean[df_clean['지역명'] == '서울'].groupby('연도')['평당분양가'].mean().plot()
```

```python
# 각 년도에 월별 분양가 box chart로 그리시오
```

```python
df_clean[df_clean.연도 == 2021].groupby('월')['평당분양가'].mean().plot(kind = 'box')
```
