```python
data = {'Team':['Riders','Riders','Devils','Devils','Kings','Giant',\

               'Kings','Kings','Riders','Royals','Royals','Riders'],\

       'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],\

       'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],\

       'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
```

```python
# data를 dataframe df로 만드시오..

df = pd.DataFrame(data)
df
```

```python
# ‘Devils’의 points 평균값은?

df.groupby('Team')['Points'].mean()
```

```python
# 팀별로 몇 개씩의 데이터가 있는가?

df.groupby('Team').size()
```

```python
# 팀별로 가장 놓은 랭킹은 어떻게 되는가?

df.groupby('Team')['Rank'].min()
```

```python
# ‘Team’의 이름을 모두 대문자로 수정하시오..

df['Team'].apply(lambda x : x.upper())
```

```python
raw_data = {

        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],

        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

df1 = pd.DataFrame(raw_data)

 

raw_data = {

        'subject_id': ['4', '5', '6', '7', '8'],

        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],

        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

df2 = pd.DataFrame(raw_data)
```

```python
df1.head()
```

```python
df2.head()
```

```python
# dataframe df1과 df2를 칼럼 ‘subject_id’를 기준으로 4가지 방식으로 병합하시오

pd.merge(df1, df2, on = 'subject_id', how = 'inner')
```

```python
pd.merge(df1, df2, on = 'subject_id', how = 'right')
```

```python
pd.merge(df1, df2, on = 'subject_id', how = 'left')
```

```python
pd.merge(df1, df2, on = 'subject_id', how = 'outer')
```

```python
# df1과 df2를 column방향으로 연결하시오..

pd.concat([df1, df2], axis = 1)
```

```python
# df1과 df2를 row방향으로 연결하시오..

pd.concat([df1, df2])
```
