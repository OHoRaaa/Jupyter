```python
df = pd.read_csv('../data/phone_data.csv', index_col = 0)
df.head(10)
```

```python
# 연도만 뽑기

df['year'] = df['month'].apply(lambda x : x.split('-')[0])
df
```

```python
# 'duraion' 정수로 만들기

df['duration'] = df['duration'].apply(lambda x : int(x))
df
```

```python
df = pd.read_csv('../data/Ecommerce Purchases')
df.head(10)
```

```python
# 고객의 email에서 email domain이 각각 몇 개씩 있는지 계산하시오
# 1. 이메일에서 도메인만 추출
# 2. 각각의 도메인이 몇 개씩 있는지 

df['Email'].apply(lambda x : x.split('@')[1]).value_counts()
```

```python
# 오전 구매건수와 오후 구매 건수를 계산하시오
# (value_counts()를 사용하는 경우와 사용하지 않는 경우 각각 구하시오)
```

```python
df.groupby('AM or PM')['Lot'].count()
```

```python
df['AM or PM'].value_counts(ascending = True)
```

```python
# 고객들이 사용하는 브라우저 종류별 갯수를 출력하시오
df['Browser Info'].apply(lambda x : x.split('/')[0]).value_counts()
```

```python
raw_data = {
    'subject_id' : ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
    'test_score' : [51, 15, 15, 61, 16, 14, 15, 15, 16, 61]
}

df_a = pd.DataFrame(raw_data)
df_a
```

```python
raw_data = {
    'subject_id' : ['4', '5', '6', '7', '8'],
    'first_name' : ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name' : ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']
}

df_b = pd.DataFrame(raw_data)
df_b
```

```python
person = pd.read_csv('../data/survey_person.csv')
site = pd.read_csv('../data/survey_site.csv')
survey = pd.read_csv('../data/survey_survey.csv')
visited = pd.read_csv('../data/survey_visited.csv')
```

```python
# site, visited을 inner merge

merged_sv = pd.merge(site, visited, left_on = 'name', right_on = 'site', how = 'inner')
merged_sv
```

```python
# person, survey를 inner merge 하기

merged_ps = pd.merge(person, survey, left_on = 'ident', right_on = 'person', how = 'inner')
merged_ps
```
