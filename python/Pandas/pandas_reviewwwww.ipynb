```python
def count_missing(vec):
    null_count = pd.insull(vec)
    return sum(null_count)
```

```python
# groupby
# 동일한 특성을 갖는 데이터끼리 데이터를 그룹으로 묶는다
# 1. 데이터를 묶는 과정 => group 생성
# 2. group별 어떤 값을 출력할지 => group별 연산과정
```

```python
df = pd.read_csv('./eda_date/gapminder.tsv', sep '\t')
df.head()
```

```python
# Transform - 그룹별 값으로 데이터값을 바꾼다
```

```python
def my_zscore(x):
    return ( x-x.mean()) / x.std()
```

```python
df.groupby('year').lifeExp.transform(my_zscore)
```
