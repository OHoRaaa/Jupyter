```python
## 베스킨라빈스 vs 던킨도너츠 의 지역별 매장수를 비교하시오



```

```python
#각각 어느 구에 매장수가 많은지 비교

df['도로명주소'].str.split()
```

```python
구 = []

for i in df['도로명주소']:
    구.append(i.split()[1])
구
```

```python
df['구'] = 구
```

```python
df
```

```python
df[df['상호명'].str.contains('베스킨|배스킨', na = False)].groupby('구').size().sort_values(ascending=False).plot(kind = 'bar')
```

```python
df[df['상호명'].str.contains('던킨', na = False)].groupby('구').size().sort_values(ascending=False).plot(kind = 'bar')
```
